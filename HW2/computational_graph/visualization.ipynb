```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from config import *
```

```python
df_com = pd.read_csv('output/ComGraph_h256_e300_b450.csv')
df_com
```

```python
def plot_loss(df):
    fig, ax = plt.subplots(2, 1, figsize=(8, 6))
    ax[0].plot(df['train_acc'], label='train_acc')
    ax[0].plot(df['val_acc'], label='val_acc')
    ax[0].legend()

    ax[1].plot(df['train_loss'], label='train_loss')
    ax[1].plot(df['val_loss'], label='val_loss')
    ax[1].legend()
    # plt.savefig('com_evaluate.png')

plot_loss(df_com.loc[:300])
```
